```r
mydata$cancer_code <- recode(mydata$status, 'M' = 1, 'B' = 0)
```

```r
# Suppose we are interested in building a model with the
# radius variabele.  Explore how this variable associated with malignancy

mydata %>%
ggplot() +
  aes( x = radius, y = status, col = status) +
  geom_jitter(height = 0.05, width = 0.01, alpha = 0.5)
```

```r
mydata %>%
  ggplot()+
  aes( x = radius, y = cancer_code) +
  geom_jitter(aes(col=status), height = 0.05, width = 0.01, alpha = 0.5)+
  geom_smooth(method = 'glm', method.args = list(family=binomial), se = FALSE, col='red')

log_model <- glm(cancer_code~radius, family=binomial, data=mydata)
summary(log_model)
```

These are LaTex notations:

$ \hat p = \frac{e^{-15+1.03{radius}}}{1+e^{-15+1.03{radius}}} $

where $\hat p$ is the predicted probability of cancer using this model.

How to interpret the 'slope'?

First, acknowledge the sign. Since the coefficient is positive, we can say that overall, an increase in radius results in a correlated increase in the predicted probability of cancer.

```r
mynewdata <- data.frame(radius = c(12,13,17,18))

predicted_prob <- predict(log_model, newdata= mynewdata, type='response')

predicted_prob
```
